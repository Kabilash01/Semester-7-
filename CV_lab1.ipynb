```python
#EXP 4 - Perform the following task:
#a.	Load an image and print its properties (dimensions, channels, data type).
#b.	Resize the image to a different resolution.
#c.	Convert the image to grayscale.
#d.	Save the processed images.

import cv2
from google.colab import files
from google.colab.patches import cv2_imshow

# Upload image
uploaded = files.upload()

# Get uploaded file name
image_name = list(uploaded.keys())[0]

# Read image
image = cv2.imread(image_name)

# Display original image
print("Original Image")
cv2_imshow(image)

# Print image properties
print("Image Shape :", image.shape)
print("Image Data Type :", image.dtype)

# Resize image
resized_image = cv2.resize(image, (300, 200))
print("Resized Image")
cv2_imshow(resized_image)
# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print("Grayscale Image")
cv2_imshow(gray_image)
cv2.imwrite("resized_image.jpg", resized_image)
cv2.imwrite("gray_image.jpg", gray_image)

print("Processed image success")
```

```python
#EXP 6 -Image Manipulation: Use OpenCV functions to apply different image manipulations
#(e.g., rotation, flipping).

import cv2
from google.colab import files
from google.colab.patches import cv2_imshow
uploaded = files.upload()
image_name = list(uploaded.keys())[0]
image = cv2.imread(image_name)
cv2_imshow(image)

# Rotate image (90 degrees clockwise)
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

print("Rotated Image")
cv2_imshow(rotated_image)

# Flip image horizontally
flip_horizontal = cv2.flip(image, 1)

print("Horizontally Flipped Image")
cv2_imshow(flip_horizontal)

# Flip image vertically
flip_vertical = cv2.flip(image, 0)

print("Vertically Flipped Image")
cv2_imshow(flip_vertical)

# Save images
cv2.imwrite("rotated_image.jpg", rotated_image)
cv2.imwrite("horizontal_flip.jpg", flip_horizontal)
cv2.imwrite("vertical_flip.jpg", flip_vertical)
```
