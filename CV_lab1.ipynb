```python
#Color to gray scale img
input_image = cv2.imread('/content/dog.jfif')
grey_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_image)
```

```python
#Resize of the Image (Option 1)
new_width=128
new_height=128
resized_image=cv2.resize(grey_image,(new_width,new_height))
cv2_imshow(resized_image)
```

```python
w_ratio=0.25
h_ratio=0.75
resized_image=cv2.resize(grey_image,(0,0),grey_image,w_ratio,h_ratio)
cv2_imshow(resized_image)
```
